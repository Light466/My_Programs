```python
# Experiment 1
# Download the House pricing dataset from kaggle and map the value to Aesthetics.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("House_pricing.csv")
```

```python
#Printing the dataset
print(df)
```

```python
sns.boxplot(df)
```

```python
df.boxplot()
```

```python
sns.lineplot(df)
```

```python
sns.scatterplot(df)
```

```python
sns.scatterplot(y='area',x='price',data=df)
plt.title("Scatter Plot showing Area and Price of house")
plt.ylabel("Area of the House")
plt.xlabel("Price of the House")
```

```python
sns.scatterplot(x='area',y='price',data=df)
plt.title("Scatter Plot showing Area and Price of house")
plt.xlabel("Area of the House")
plt.ylabel("Price of the House")
```

```python
sns.scatterplot(x='area',y='price',data=df,hue=df["mainroad"])
plt.title("Area and Price of house with Mainroad")
plt.xlabel("Area of the House")
plt.ylabel("Price of the House")
```

```python
sns.lineplot(x='area',y='price',data=df,hue=df["mainroad"])
plt.title("Area and Price of house with Mainroad")
plt.xlabel("Area of the House")
plt.ylabel("Price of the House")
```
